# Text Preprocessing Steps

In this section, I will go through several key steps for text preprocessing. I will be cleaning and transforming text data into a format that is easier for models to process.

## 1. Remove URLs
URLs can often be present in text data, especially in content sourced from web pages, social media, or forums. Since URLs do not provide meaningful information for text analysis or natural language processing tasks, I will remove them to reduce noise.

## 2. Remove Emails
Emails, similar to URLs, can appear in text data and are generally not useful for text analysis. They often serve as metadata rather than content. Removing email addresses helps in cleaning the text further.

## 3. Remove Special Characters
Special characters such as punctuation marks, symbols, and numbers may not contribute to understanding the context or meaning of textual content. Removing these characters helps in simplifying the text and focusing on the words themselves.

## 4. Remove Stop Words
Stop words are commonly used words (such as "is", "the", "in") that are often filtered out before processing text data because they do not carry significant meaning or contribute to the analysis. Removing stop words helps in focusing on the more meaningful words that contribute to understanding the context.

By performing these preprocessing steps, I can clean our text data and make it more suitable for analysis or machine learning tasks.

```python
def clean(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    email_pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b')
    clean_text = url_pattern.sub('', text)
    clean_text = email_pattern.sub('', clean_text)
    
    clean_text = re.sub('[^\w\s]', '', clean_text)
    stop_words = set(stopwords.words('english'))
    clean_text  = ' '.join(word for word in clean_text.split() if word.lower() not in stop_words)
    
    return clean_text
```

# TfIdfVectorizer

`tfidf = TfidfVectorizer()`:
This line initializes an instance of the TfidfVectorizer class. TfidfVectorizer is a tool that converts a collection of raw documents (text data) into a matrix of TF-IDF features.

`TF (Term Frequency)`: Measures how frequently a term (word) appears in a document. <br>
`IDF (Inverse Document Frequency)`: Measures how important a term is. This is calculated by taking the logarithm of the total number of documents divided by the number of documents containing the term. Words that appear frequently across many documents (like "the", "is", etc.) are down-weighted by the IDF calculation.
The resulting TF-IDF score is a product of these two metrics. This helps to normalize the term frequency by how rarely the term appears in the entire dataset, thus giving more weight to more "important" terms.

```python
tfidf = TfidfVectorizer()
tfidf.fit(df['Cleaned_Resume'])
resume = tfidf.transform(df['Cleaned_Resume'])
```

# Train Test Split

```python
X_train , X_test, y_train , y_test = train_test_split(resume, df['Category'], test_size = 0.2, random_state = 42)
```

# Model Training

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
```

```python
models = {
    'KNeighborsClassifier':KNeighborsClassifier(),
    'LogisticRegression': LogisticRegression(),
    'RandomForestClassifier': RandomForestClassifier(),
    'SVC': SVC(),
    'MultinomialNB': MultinomialNB(),
}
```

```python
accuracy_scores = {}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy  = accuracy_score(y_test, y_pred)
    accuracy_scores[model_name] = accuracy
    print(f"Accuarcy of {model_name} on test data: {accuracy}")
```
